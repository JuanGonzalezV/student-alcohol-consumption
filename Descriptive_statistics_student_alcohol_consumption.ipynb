```python
#Step 1
# The dataset is about the age of absences of a sample population by gender
# Read the file data.csv into a datframe

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from datasets import load_dataset
```

```python
ds = load_dataset("scikit-learn/student-alcohol-consumption")
```

```python
df = pd.DataFrame.from_dict(ds['train'])
df
```

```python
df.to_csv('student-alcohol-consumption.csv', index=False)
```

Format columns

```python
df = df[["sex","age","absences"]]
df = df.rename(columns={'sex': 'gender'})
df['gender'] = df['gender'].replace({'F': 'female', 'M': 'male'})
df
```

```python
#Step 2
# Print the summary statistics overall
print(df.describe())
```

```python
#Step 3
# Print the summary statistics by gender for age
print(df.groupby('gender')['age'].describe())
```

```python
#Step 4
# Print the summary statistics by gender for age
print(df.groupby('gender')['absences'].describe())
```

```python
#Step 5
#Plot the histogram of age and the histogram of age by gender
sns.histplot(df['age'], kde=True)
plt.show()
sns.histplot(data=df, x="age", hue="gender")
plt.show()
```

```python
#Step 6
# Plot the histogram of absences and the absences by gender

sns.histplot(df['absences'], kde=True)
plt.show()
sns.histplot(data=df, x="absences", hue="gender")
plt.show()
```

```python
#Step 7
#Plot the boxplot of absences by gender

import seaborn as sns
sns.boxplot(x = 'gender', y = 'absences', data = df, hue='gender')
```

```python
#Step 8
#Plot the boxplot of the age by gender

sns.boxplot(x = 'gender', y = 'age', data = df, hue='gender')
```
